```python
import pandas as pd
import numpy as nm
df = pd.read_excel("C:\\Users\\artem\\YandexDisk\\Porg\\Py\\stsystem\\direct.xlsx")
df.head(4)
```

```python
def convet_price_rate (df):
    if 0 < df.cost < 400:
        return "Wery well"
    elif 400 < df.cost < 700:
        return "Normal"
    elif df.cost == 0:
        return "-"
    else:
        return "Wery bad"
```

```python
rating_converse = df.apply(convet_price_rate, axis='columns')
df['rating_converse '] = rating_converse
df.head()
```
